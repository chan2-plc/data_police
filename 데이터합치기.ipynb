```python
cp1 = pd.DataFrame({
    '연도': ['2018', '2019', '2020', '2021', '2022'],
    '소비자물가 총지수(2020=100)': [99.1, 99.5, 100.0, 102.5, 107.7],
    '소비자물가상승률(%)': [1.5, 0.4, 0.5, 2.5, 5.1]
})

df1 = pd.DataFrame({
    '연도': ['2018', '2019', '2020', '2021', '2022'],
    '전자금융 발생건수[건]': [7877.0, 8132.0, 6402.0, 7024.0, 9731.0]
})

# '연도' 열을 기준으로 두 데이터프레임을 합침
joined_df = pd.merge(cp1, df1, on='연도', how='outer')

# 결과 DataFrame에서 '연도' 열을 맨 앞쪽으로 이동
joined_df = joined_df[['연도'] + [col for col in joined_df.columns if col != '연도']]

# 결과 출력
print(joined_df)
```
