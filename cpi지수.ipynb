```python
import requests
from bs4 import BeautifulSoup
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://kostat.go.kr/cpidtval.es?mid=b70201010000'
r=requests.post(url)
soup=BeautifulSoup(r.text, "html.parser")
```

```python
table=soup.find("table")
rows=table.find('tbody').find_all('tr')
print(table)
print(rows)
```

```python
headers = [th.text.strip() for th in table.select('thead th')[1:]]


data = {}
for row in table.select('tbody tr'):
    row_data = [td.text.strip() for td in row.select('td')]
    category = row.select_one('th').text.strip()
    data[category] = row_data


print("Headers:", headers)
for category, values in data.items():
    print(category, ":", values)
```

```python
headers = [th.text.strip() for th in table.select('thead th')[1:]]


data = {}
for row in table.select('tbody tr'):
    row_data = [td.text.strip() for td in row.select('td')]
    category = row.select_one('th').text.strip()
    data[category] = row_data


csv_file_path = r'C:\Users\juntaek\Desktop\extracted_data.csv'

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    

    writer.writerow(['Category'] + headers)
    

    for category, values in data.items():
        writer.writerow([category] + values)

print(f"CSV file saved to: {csv_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import platform

if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    plt.rc('font', family='AppleGothic')
else: #linux
    plt.rc('font', family='NanumGothic')

data = {
    '연도': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    '소비자물가총지수': [91.8, 93.0, 94.2, 94.9, 95.8, 97.6, 99.1, 99.5, 100.0, 102.5, 107.7],
    '소비자물가상승률': [2.2, 1.3, 1.3, 0.7, 1.0, 1.9, 1.5, 0.4, 0.5, 2.5, 5.1]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df['연도'], df['소비자물가총지수'], marker='o', label='소비자물가총지수')
plt.plot(df['연도'], df['소비자물가상승률'], marker='o', label='소비자물가상승률')

# 그래프 꾸미기
plt.title('소비자물가총지수와 소비자물가상승률 변화')
plt.xlabel('연도')
plt.ylabel('지수 및 상승률 (%)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
cpi_total_index = [91.8, 93.0, 94.2, 94.9, 95.8, 97.6, 99.1, 99.5, 100.0, 102.5, 107.7]
cpi_inflation_rate = [2.2, 1.3, 1.3, 0.7, 1.0, 1.9, 1.5, 0.4, 0.5, 2.5, 5.1]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('년도')
ax1.set_ylabel('소비자물가 총지수', color=color)
ax1.plot(years, cpi_total_index, color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Y축을 공유하는 두 번째 축 생성

color = 'tab:blue'
ax2.set_ylabel('소비자물가 상승률 (%)', color=color)
ax2.plot(years, cpi_inflation_rate, color=color, marker='x')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('년도별 소비자물가총지수와 소비자물가상승률')
plt.show()
```
