```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('police_2_utf8.csv')
df
```

```python
df1=df.iloc[2]
df1
```

```python
df2=pd.read_csv('소비자물가지수 .csv')
df2
```

```python
finaldf = pd.concat([df1, df2], axis=1)
finaldf
```

```python
finaldf.info()
```

```python
correlation_matrix = finaldf.corr(method='pearson', numeric_only=True)
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm')
plt.title('Correlation Matrix Heatmap for Embezzlement')
plt.xlabel('Consumer Price Index(Year)')
plt.ylabel('Consumer Price Index(Year)')
plt.show()
```

```python
corr_pairs=correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0]
high_corr
```
