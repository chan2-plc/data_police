```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cpi=pd.read_excel(r'C:\Users\juntaek\Desktop\소비자물가지수.xlsx')
cpi
```

```python
cpi.dropna(axis=0, how='any', inplace=True)
cpi
```

```python
selected_rows = cpi.iloc[0:2, :]
selected_rows
```

```python
selected_rows.index
```

```python
import matplotlib.pyplot as plt

import platform

if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    plt.rc('font', family='AppleGothic')
else: #linux
    plt.rc('font', family='NanumGothic')

years = selected_rows.columns[1:]
cpi_values = selected_rows.iloc[0, 1:].astype(float)

plt.plot(years, cpi_values, marker='o', linestyle='-', color='b')


plt.xlabel('연도')
plt.ylabel('소비자물가 총지수(2020=100)')
plt.title('연도별 소비자물가 총지수 변화')

plt.grid(True)

plt.show()
```
